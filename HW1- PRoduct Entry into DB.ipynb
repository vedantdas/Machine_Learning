```python
import pandas as pd
import sqlalchemy
from os import path

#To read the read the data from excel sheet.
df = pd.read_csv('D:\My Git\Xlsx Files\Prod_Details.csv')
df.head()

#To create a connection with the database.
engine = sqlalchemy.create_engine('postgresql+psycopg2://postgres:codyhbk1@127.0.0.1:5432/firstdb')
print (engine.table_names())

#To push the data from the excel sheet to the server database.
df.to_sql('Prod_Details', index = True,index_label=None, con=engine, if_exists = 'append')
print ("Push to server successful")

#To fetch the data from the database.
pdf = pd.read_sql_query('select * from "Prod_Details"',con=engine)
pdf.head()
```
